```python

import os
import typer
from typing import Optional
from rich.prompt import Prompt

from phi.agent import Agent
from phi.knowledge.pdf import PDFUrlKnowledgeBase
from phi.vectordb.pineconedb import PineconeDB

#api_key = os.getenv("PINECONE_API_KEY")
index_name = "thai-recipe-hybrid-search"

vector_db = PineconeDB(
    name=index_name,
    dimension=768,
    metric="cosine",
    spec={"serverless": {"cloud": "aws", "region": "us-east-1"}},
    api_key=api_key,
    use_hybrid_search=True,
    hybrid_alpha=0.5,
    embedder=GeminiEmbedder()
)

knowledge_base = PDFUrlKnowledgeBase(
    urls=["https://phi-public.s3.amazonaws.com/recipes/ThaiRecipes.pdf"],
    vector_db=vector_db,
)

# Comment out after first run
knowledge_base.load(recreate=True, upsert=True)


def pinecone_agent(user: str = "user"):
    run_id: Optional[str] = None

    agent = Agent(
        run_id=run_id,
        user_id=user,
        model=Gemini(id="gemini-1.5-flash"),
        knowledge=knowledge_base,
        search_knowledge=True,
        tools=[DuckDuckGo()],
        show_tool_calls=True,
        debug_mode=True,
    )

    if run_id is None:
        run_id = agent.run_id
        print(f"Started Run: {run_id}\n")
    else:
        print(f"Continuing Run: {run_id}\n")

    while True:
        message = Prompt.ask(f"[bold] :sunglasses: {user} [/bold]")
        if message in ("exit", "bye"):
            break
        agent.print_response(message)

# Instead of using typer.run, call the function directly:
if __name__ == "__main__":
    pinecone_agent()  # Call the function directly
#if __name__ == "__main__":
   # typer.run(pinecone_agent)
```

```python

from typing import Optional

import typer
from agno.agent import Agent
from agno.knowledge.pdf_url import PDFUrlKnowledgeBase
from agno.vectordb.lancedb import LanceDb
from agno.vectordb.search import SearchType
from rich.prompt import Prompt

# LanceDB Vector DB
vector_db = LanceDb(
    table_name="recipes",
    uri="tmp/lancedb",
    search_type=SearchType.hybrid,
    embedder=GeminiEmbedder()
)

# Knowledge Base
knowledge_base = PDFUrlKnowledgeBase(
    urls=["https://agno-public.s3.amazonaws.com/recipes/ThaiRecipes.pdf"],
    vector_db=vector_db,
)


def lancedb_agent(user: str = "user"):
    agent = Agent(
        user_id=user,

        model=Gemini(id="gemini-1.5-flash"),        knowledge=knowledge_base,
        search_knowledge=True,
        show_tool_calls=True,
        debug_mode=True,
    )

    while True:
        message = Prompt.ask(f"[bold] :sunglasses: {user} [/bold]")
        if message in ("exit", "bye"):
            break
        agent.print_response(message)

if __name__ == "__main__":
    # Comment out after first run
    knowledge_base.load(recreate=False)

    typer.run(lancedb_agent)
```

## Query sqlite

```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect('agents_rag.db')

# Create a cursor object
cursor = conn.cursor()

# Execute a query to retrieve all records from the 'phidata' table
cursor.execute('SELECT * FROM phidata')

# Fetch all rows from the executed query
rows = cursor.fetchall()

# Process and print the results
for row in rows:
    print(row)

# Close the cursor and connection
cursor.close()
conn.close()
```
